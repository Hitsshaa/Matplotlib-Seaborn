# Barplot

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue,label = "Revenue")
plt.legend()
```

```python
plt.bar(company,revenue,label = "Revenue",width = .2,color ="r")
plt.legend()
```

```python
company_as_number  = np.arange(len(company))   # if we want to give values in number
company_as_number
```

```python
plt.bar(company_as_number,revenue,label = "Revenue")
plt.legend()
```

```python
plt.bar(company_as_number,revenue,label = "Revenue",linewidth = 10,edgecolor = "y")
plt.xticks(company_as_number,company)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company report")
style.use("ggplot")   # use to change background color
```

```python
Multiple Bars showing revenue and profit of major US tech companies
```

```python
profit = [23,67,45,5]
```

```python
plt.bar(company_as_number,revenue,label = "Revenue",width = .2)   # technical it is wrong because d/f betweeen graph
plt.bar(company_as_number,profit,width = .2)
plt.xticks(company_as_number,company)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company report")
```

```python
plt.bar(company_as_number-.2,revenue,label = "Revenue",width = .2)   # technical it is wrong because d/f betweeen graph
plt.bar(company_as_number+.2,profit,width = .2)
plt.xticks(company_as_number,company)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company report")
```

```python
plt.bar(company_as_number-.1,revenue,label = "Revenue",width = .2)
plt.bar(company_as_number+.1,profit,width = .2)
plt.xticks(company_as_number,company,rotation =60)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us top stock")
```

```python
width = .2
plt.bar(company_as_number,revenue,label = "Revenue",width = .2 )
plt.bar(company_as_number+width,profit,width = .2 )
plt.xticks(company_as_number,company,rotation =60)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us top stock")
plt.show()
```

```python
Horizontal bar chart using barh function
```

```python
plt.barh(company_as_number,revenue,label = "Revenue" )

plt.xticks(company_as_number,company)   # give string value back to plot
plt.legend()
plt.xlabel("company")
plt.ylabel("revenue")
plt.title("us top stock")
```
